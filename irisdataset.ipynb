```python
plt.figure(figsize=(8,2))
ax = sns.countplot(data=iris_data, x=iris_data["Species"],palette="Spectral")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.title("Species Count")
plt.xlabel('species')
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10,8))
sns.scatterplot(x=iris_data['SepalLengthCm'],y=iris_data['SepalWidthCm'], data=iris_data[0:2000],alpha=1,color='black')
plt.title('Comparison between various species based on sapel length and width',fontsize=15)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10,8))
sns.scatterplot(x=iris_data['PetalLengthCm'],y=iris_data['PetalWidthCm'], data=iris_data[0:2000],alpha=1)
plt.title('Comparison between various species based on sapel length and width',fontsize=15)
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.boxplot( y='PetalWidthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[0, 0], patch_artist=True, boxprops = dict(facecolor = "lightblue"), medianprops = dict(color = "darkblue", linewidth = 1.5), capprops = dict(color = "darkblue", linewidth = 2))
sns.boxplot( y='PetalLengthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[0, 1], patch_artist=True, boxprops = dict(facecolor = "lightgreen"), medianprops = dict(color = "darkgreen", linewidth = 1.5), capprops = dict(color = "darkgreen", linewidth = 2))
sns.boxplot( y='SepalLengthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[1, 0], patch_artist=True, boxprops = dict(facecolor = "tan"), medianprops = dict(color = "brown", linewidth = 1.5), capprops = dict(color = "brown", linewidth = 2))
sns.boxplot( y='SepalWidthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[1, 1], patch_artist=True, boxprops = dict(facecolor = "violet"), medianprops = dict(color = "purple", linewidth = 1.5), capprops = dict(color = "purple", linewidth = 2))
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,10))
sns.violinplot( y='PetalWidthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[0, 0],inner='quartile', color="palevioletred")
sns.violinplot( y='PetalWidthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[0, 1],inner='quartile', color="peachpuff")
sns.violinplot( y='SepalLengthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[1, 0],inner='quartile', color="lightslategrey")
sns.violinplot( y='SepalWidthCm', x= 'Species', data=iris_data, orient='v' , ax=axes[1, 1],inner='quartile', color="powderblue")
plt.show()
```

```python
labels = {
    0: "Setosa",
    1: "Versicolor ",
    2: "Virginica"
}

values = {
    0: 100,
    1: 320,
    2: 291
}
```

```python
iris_data = pd.DataFrame(values.values(), index=values.keys(), columns=['Species'])

# Ploting the pie chart
# Use specific colors for each label
explode = (0.1, 0, 0)  # Explode the first slice for emphasis

plt.pie(iris_data['Species'], labels=[labels[key] for key in iris_data.index], autopct='%1.1f%%', startangle=360, explode=explode)
plt.title("Distribution of Iris Data")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
